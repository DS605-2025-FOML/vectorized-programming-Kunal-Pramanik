### 2. Data Cleaning and Preprocessing: 

>> The Cabin column has many missing values. Drop this column entirely from the DataFrame. 

```python
df1 = df.drop(columns=['Cabin'])
df1
```

>>Fill the missing values in the Age column using the mean age.

```python
df1['Age']=df1['Age'].fillna(df1['Age'].mean())
df1
```

>>The Embarked column has a few missing values. Fill them with the mode (the most common 
port of embarkation). 

```python
mode_embarked = df1['Embarked'].mode()[0] # [0] to get the first mode value as if there was a tie then it gives more than one in series
#mode_embarked
df1['Embarked'] = df1['Embarked'].fillna(mode_embarked)
df1
```

>> Confirm that your DataFrame now has no missing values in any of the remaining columns. 

```python
df1_null=df1.isnull()
df1_null.sum() # .sum() is used to get the total count of missing values in each column
```

```python
df1_null.values.any()  # to check if there is any NaN value in the entire dataframe
```

### 3. Feature Engineering: 

>>Create a new column called FamilySize which is the sum of the SibSp and Parch columns, plus 
1 (to count the passenger themselves). 

```python
df2=df1.copy()
df2
```

```python
df2["FamilySize"] = df2["SibSp"] + df2["Parch"] + 1
df2
```

>> Create a categorical column called AgeGroup. Bin the Age column into the following groups: 
â–  Child: 0-12 
â–  Teen: 13-19 
â–  Adult: 20-59 
â–  Senior: 60+ 

```python
bins=[0,12,19,59,np.inf]
labels=['Child','Teen','Adult','Senior']
df2['AgeGroup'] = pd.cut(df2['Age'], bins=bins, labels=labels, right=True)
```

```python
df2
```

### 5. Grouping and Aggregation (Groupby): 

>>What was the average age of passengers who survived versus those who did not? (Hint: Group
